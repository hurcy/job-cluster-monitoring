# Cluster Policy Names 테이블 갱신

Databricks SDK로 워크스페이스의 클러스터 정책 목록을 가져와 `hurcy.default.cluster_policy_names` 테이블에 저장한다.

```python
from databricks.sdk import WorkspaceClient

w = WorkspaceClient()
```

```python
policies = [
    {"policy_id": p.policy_id, "policy_name": p.name}
    for p in w.cluster_policies.list()
]
print(f"Found {len(policies)} policies")
```

```python
catalog = "hurcy"
schema = "default"
table_name = f"{catalog}.{schema}.cluster_policy_names"
```

```python
df = spark.createDataFrame(policies, schema="policy_id STRING, policy_name STRING")
df.write.mode("overwrite").saveAsTable(table_name)
print(f"OK: {len(policies)} policies → {table_name}")
```
